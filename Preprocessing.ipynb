```python
import re

def clean_ingredients(ingredients):
    cleaned_ingredients = []
    
    for ingredient in ingredients:
        # Menghapus angka dan ukuran dari tiap bahan makanan
        ingredient = re.sub(r'\d+(\.\d+)?\s*\w+', '', ingredient)
        
        # Menghapus tanda baca
        ingredient = re.sub(r'[^\w\s]', '', ingredient)
        
        # Mengubah huruf menjadi huruf kecil
        ingredient = ingredient.lower()
        
        # Menghapus kata-kata yang tidak relevan
        stop_words = [
    '10', '100', '1000', '11', '12', '13', '15', '150', '1500', '1sdt',
    '1sendok', '20', '200', '20ml', '25', '250', '2cm', '30', '300',
    '350', '3cm', '400', '50', '500', '50ml', '60', '600', '65',
    '65ml', '700', '750', '7bh', '7buah', '800ml',  'ada', 'adas',
    'agak', 'aja', 'aka','akuu', 'ambil','apa', 'api', 'aq', 'asam',
    'asin', 'atau', 'atau3', 'b51', 'badan','bag', 'bagian', 'bagus', 'bahan',
    'bakar','balado', 'baluran', 'bamer','banyak', 'baput', 'baru', 'basah', 'batang', 'batu',
    'bawah', 'belah', 'berapa','beri', 'berkulit', 'bersih', 'bersihkan', 'besar', 'bh', 'biar',
    'biasa', 'bijinya', 'bikin', 'bila', 'biru',
    'bisa', 'bks', 'blok',  'boleh','bongkol', 'botolan', 'brand',
    'brenebron', 'btg', 'btng', 'btr', 'bu',
    'buah', 'buang', 'buat', 'bubuk', 'bulat', 'bulatan', 'bumbu',
    'bumbunya', 'bunga', 'bungkus', 'butir',
    'butiran',  'cacah', 'cair',
    'caisim', 'cambah', 'campur', 'campuran','ceplok', 'cincang', 'cm', 'cookpad',
    'cubes', 'cuci', 'cuka', 'cup', 'dadu',
    'dahulu', 'dalam', 'dan', 'dari', 'dasar', 'dati', 'daun',
    'daunnya', 'dengan', 'di', 'diabetasol',
    'diameter', 'diblender', 'dicairkan', 'diganti', 'digeprak',
    'digeprek', 'dihalus']
        ingredient = ' '.join([word for word in ingredient.split() if word not in stop_words])
        
        # Menambahkan bahan makanan yang telah dibersihkan ke dalam daftar hasil
        if ingredient:
            cleaned_ingredients.append(ingredient)
    
    return cleaned_ingredients
```

```python
df['Parser Ingredient'] = clean_ingredients(df['Ingredients'])

# Menampilkan DataFrame hasil
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
 
# Inisialisasi TfidfVectorizer
tf = TfidfVectorizer()
 
# Melakukan perhitungan idf pada data cuisine
tf.fit(df['Parser Ingredient']) 
 
# Mapping array dari fitur index integer ke fitur nama
tf.get_feature_names_out() 
```

```python
# Melakukan fit lalu ditransformasikan ke bentuk matrix
tfidf_matrix = tf.fit_transform(df['Ingredients']) 
 
# Melihat ukuran matrix tfidf
tfidf_matrix.shape 
```

```python
# Mengubah vektor tf-idf dalam bentuk matriks dengan fungsi todense()
tfidf_matrix.todense()
```

```python
pd.DataFrame(
    tfidf_matrix.todense(), 
    columns=tf.get_feature_names_out(),
    index=df.Title
).sample(626, axis=1).sample(10, axis=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
 
# Menghitung cosine similarity pada matrix tf-idf
cosine_sim = cosine_similarity(tfidf_matrix) 
cosine_sim
```

```python
# Membuat dataframe dari variabel cosine_sim dengan baris dan kolom berupa nama resto
cosine_sim_df = pd.DataFrame(cosine_sim, index=df['Title'], columns=df['Title'])
print('Shape:', cosine_sim_df.shape)
 
# Melihat similarity matrix pada setiap resep
cosine_sim_df.sample(5, axis=1).sample(10, axis=0)
```

```python
def resep_recommendations(Bahan, similarity_data=cosine_sim_df, items=df, k=10):
    """
    Rekomendasi Resto berdasarkan kemiripan dataframe
 
    Parameter:
    ---
    nama_resto : tipe data string (str)
                Nama Restoran (index kemiripan dataframe)
    similarity_data : tipe data pd.DataFrame (object)
                      Kesamaan dataframe, simetrik, dengan resto sebagai 
                      indeks dan kolom
    items : tipe data pd.DataFrame (object)
            Mengandung kedua nama dan fitur lainnya yang digunakan untuk mendefinisikan kemiripan
    k : tipe data integer (int)
        Banyaknya jumlah rekomendasi yang diberikan
    ---
 
 
    Pada index ini, kita mengambil k dengan nilai similarity terbesar 
    pada index matrix yang diberikan (i).
    """
 
 
    # Mengambil data dengan menggunakan argpartition untuk melakukan partisi secara tidak langsung sepanjang sumbu yang diberikan    
    # Dataframe diubah menjadi numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:,Bahan].to_numpy().argpartition(
        range(-1, -k, -1))
    
    # Mengambil data dengan similarity terbesar dari index yang ada
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    # Drop nama_resto agar nama resto yang dicari tidak muncul dalam daftar rekomendasi
    closest = closest.drop(Bahan, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```
